```python
import numpy as np
from qiskit_nature.drivers import UnitsType, Molecule
from qiskit_nature.second_q.drivers import PySCFDriver
from qiskit_nature.second_q.mappers import BravyiKitaevMapper
from qiskit_nature.second_q.mappers import QubitConverter
```

```python
# ---------------------- For the full System  -----------------------

length = np.arange(0.2,3,0.5)
out=[]

for bond_length in length:
    driver = PySCFDriver(atom='H 0.0 0.0 0.0; H 0.0 0.0 '+str(bond_length), unit=UnitsType.ANGSTROM, basis='sto3g')
    problem = driver.run()
    fermionic_op = problem.hamiltonian.second_q_op()
    mapper = BravyiKitaevMapper()
    qubit_parity_op = mapper.map(fermionic_op)
    out += [qubit_parity_op]
    print(out)
    
```

```python
# ---------------------- Two qubit Reduction  -----------------------

length = np.arange(0.2,3,0.5)
out=[]

for bond_length in length:
    driver = PySCFDriver(atom='H 0.0 0.0 0.0; H 0.0 0.0 '+str(bond_length), unit=UnitsType.ANGSTROM, basis='sto3g')
    problem = driver.run()
    mapper = BravyiKitaevMapper()
    converter = QubitConverter(BravyiKitaevMapper(), two_qubit_reduction=True)
    qubit_op = converter.convert(fermionic_op, num_particles=problem.num_particles)
    out += [qubit_op]
    print(out)
```
